```python
import directlfq.config as lfq_config
import directlfq.normalization as lfqnorm
import directlfq.protein_intensity_estimation as lfqprot_estimation
import directlfq.utils as lfqutils


INPUT_FILE = "../../test_data/system_tests/quicktests/diann/shortened_input.tsv.diann_fragion_isotopes_raw.aq_reformat.tsv"


group_column = "pg"
lfq_df = lfqutils.import_data(input_file=INPUT_FILE, input_type_to_use="directlfq", filter_dict={})
lfq_df = lfq_df.rename(columns={"protein": group_column})
lfq_df = lfq_df.sort_values(by=[group_column], ignore_index=True)
display(lfq_df)


lfq_config.set_global_protein_and_ion_id(protein_id=group_column, quant_id="ion")
lfq_config.set_log_processed_proteins(log_processed_proteins=True)
lfq_config.set_compile_normalized_ion_table(compile_normalized_ion_table=True)



lfq_config.set_global_protein_and_ion_id(protein_id="pg", quant_id="ion")

lfq_df = lfqutils.index_and_log_transform_input_df(lfq_df)
lfq_df = lfqutils.remove_allnan_rows_input_df(lfq_df)

lfq_df = lfqnorm.NormalizationManagerSamplesOnSelectedProteins(lfq_df, num_samples_quadratic=20, selected_proteins_file=None).complete_dataframe

protein_df, ion_df = lfqprot_estimation.estimate_protein_intensities(lfq_df,min_nonan=1,num_samples_quadratic=20, num_cores = 2)

display(protein_df)
```
