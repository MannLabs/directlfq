```python
input_file = "../../test_data/system_tests/lfqbench/DIANN/report.tsv"
maxquant_file = "../../test_data/system_tests/lfqbench/maxDIA/proteinGroups.txt"
diann_file = "../../test_data/system_tests/lfqbench/DIANN/report.tsv"
diann_lfq_file = "../../test_data/system_tests/lfqbench/DIANN/report.tsv.diann_protein.aq_reformat.tsv"

directlfq_output = "../../test_data/system_tests/lfqbench/DIANN/report.tsv.protein_intensities.tsv"

samplemap_directlfq = "../../test_data/system_tests/lfqbench/DIANN/samplemap.tsv"
samplemap_diann = "../../test_data/system_tests/lfqbench/DIANN/samplemap.tsv"

samplemap_maxquant = "../../test_data/system_tests/lfqbench/maxDIA/samplemap.maxLFQ.tsv"

run_directlfq = True
```

```python
if run_directlfq:
    import directlfq.lfq_manager as lfqmgr

    lfqmgr.run_lfq(input_file=input_file, min_nonan=7, input_type_to_use='diann_precursors')

```

```python
import directlfq.benchmarking as lfqbenchmark
import directlfq.utils as lfqutils

samplemap_df_directlfq = lfqutils.load_samplemap(samplemap_directlfq)
samples_used_directlfq = lfqutils.get_samples_used_from_samplemap_df(samplemap_df_directlfq, cond1="A", cond2="B")
samplemap_df_maxlfq = lfqutils.load_samplemap(samplemap_maxquant)
samples_used_maxlfq = lfqutils.get_samples_used_from_samplemap_df(samplemap_df_maxlfq, cond1="A", cond2 = "B")


import pandas as pd

df = pd.read_csv(directlfq_output, sep = "\t")
display(df)
restable_directlfq = lfqbenchmark.ResultsTableDirectLFQ(input_file=directlfq_output, input_name="directLFQ", samples_c1=samples_used_directlfq[0], samples_c2=samples_used_directlfq[1])
restable_diannlfq = lfqbenchmark.ResultsTableDirectLFQ(input_file=diann_lfq_file, input_name="diannLFQ", samples_c1=samples_used_directlfq[0], samples_c2=samples_used_directlfq[1])
restable_maxlfq = lfqbenchmark.ResultsTableMaxQuant(input_file=maxquant_file, input_name="maxLFQ", samples_c1=samples_used_maxlfq[0], samples_c2=samples_used_maxlfq[1])
```

```python
display(restable_directlfq.formated_dataframe)
display(restable_diannlfq.formated_dataframe)
display(restable_maxlfq.formated_dataframe)
```

```python
organism_annotator_mq = lfqbenchmark.OrganismAnnotatorMaxQuant(mapping_file=maxquant_file)
display(restable_maxlfq.formated_dataframe)
organism_annotator_mq.annotate_table_with_organism(restable_maxlfq)
display(restable_maxlfq.formated_dataframe)
```

```python
organism_annotator_diannlfq = lfqbenchmark.OrganismAnnotatorDIANN(mapping_file=diann_file,protein_column='Protein.Group')
organism_annotator_diannlfq.annotate_table_with_organism(restable_diannlfq)
```

```python
organism_annotator_diann = lfqbenchmark.OrganismAnnotatorDIANN(mapping_file=diann_file, protein_column='Protein.Group')

organism_annotator_diann.annotate_table_with_organism(restable_directlfq)
```
