```python
DIANN_FILE = "../../test_data/system_tests/diaPASEF_diann/report.tsv"
```

```python
import pandas as pd
df_diann = pd.read_csv(DIANN_FILE, sep="\t")
```

```python
display(df_diann)
display(df_diann.columns)
```

```python
peptide2protgroups = {}
for i, row in df_diann.iterrows():
	peptide = row["Stripped.Sequence"]
	protgroup = row["Protein.Group"]
	if peptide not in peptide2protgroups:
		peptide2protgroups[peptide] = set()
	peptide2protgroups[peptide].add(protgroup)


    
```

```python
for protgroups in peptide2protgroups.values():
	assert len(protgroups) == 1
	print("passed")
```
